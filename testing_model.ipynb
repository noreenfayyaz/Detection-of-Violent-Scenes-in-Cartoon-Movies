```python
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import BatchNormalization
from keras.models import Sequential
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import VGG16

#model architectures
model = Sequential()
model.add(LSTM(32,dropout=0.5, recurrent_dropout=0.5, return_sequences=True,input_shape=(49,512)))  # returns a sequence of vectors of dimension 32
model.add(LSTM(32, return_sequences=True))  # returns a sequence of vectors of dimension 32
model.add(LSTM(32))# return a single vector of dimension 32
model.add(BatchNormalization())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
import cv2
import numpy as np


# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture("test_vid/normal.mpg")
target = ['violence','Normal']

# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")

# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frame
  ret, frame = cap.read()
    
  if ret == True:
    
    # draw the label into the frame
    #__draw_label(frame, 'Hello World', (20,20), (255,0,0))
    frame1=cv2.resize(frame,(224,224))
    #CONVERT TO NUMPY ARRAY
    #x = image.img_to_array(frame)
    x = np.expand_dims(frame1, axis=0)
    #PREPARE THR IMAGE FOR THE VGG MODEL
    x = preprocess_input(x)
    #print('Input image shape:', x.shape)
    
    img_data = np.array(x) 
    #print (img_data.shape)

    #LOAD MODEL
    model = VGG16(weights='imagenet', include_top=False) 
    # get extracted features of test_set
    x_test = model.predict(x)
    #print(x_test.shape)
    #reshape 4D to 2D
    x_test= x_test.reshape(x_test.shape[0],x_test.shape[1] * x_test.shape[2],x_test.shape[3])
    print(x_test.shape)
    # font 
    font = cv2.FONT_HERSHEY_SIMPLEX 
  
    # org 
    org = (50, 50) 
  
    # fontScale 
    fontScale = 1
   
    #Blue color in BGR 
    color = (0, 0, 255) 
  
    # Line thickness of 2 px 
    thickness = 2
    model = Sequential()
    model.add(LSTM(32,dropout=0.5, recurrent_dropout=0.5, return_sequences=True,input_shape=(49,512)))  # returns a sequence of vectors of dimension 32
    model.add(LSTM(32, return_sequences=True))  # returns a sequence of vectors of dimension 32
    model.add(LSTM(32))# return a single vector of dimension 32
    model.add(BatchNormalization())
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    
    
    #result = model.predict_classes(x_test)
    #result = np.argmax(result)
    #result = target[np.argmax(model.predict_classes(x_test))]
    result =np.argmax(model.predict(x_test),axis=1)
    result=str(result)
    print(result)
    #if result == 0:        
        #image = cv2.putText(frame,'Anamoly_Detected', org, font, fontScale, 
        #color, thickness, cv2.LINE_AA, False)
    image = cv2.putText(frame,  result, org, font, fontScale,  color, thickness, cv2.LINE_AA, False)  


    # Display the resulting frame
    cv2.imshow('Frame',frame)

    # Press Q on keyboard to  exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
      break

  # Break the loop
  else: 
    break

# When everything done, release the video capture object
cap.release()

# Closes all the frames
cv2.destroyAllWindows(0)
```
